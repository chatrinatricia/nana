Set adalah tipe data koleksi yang tidak berurutan (unordered), tidak terindeks, dan hanya menyimpan elemen unik (otomatis menghapus duplikat). Set sangat berguna untuk operasi himpunan seperti irisan, gabungan, dan selisih. Dibuat menggunakan kurung kurawal {} atau fungsi set(), set bersifat mutable (elemen bisa ditambah/dihapus).

```python
s = set([3, 4, 9, 10])  
t = set("Hello")        

print(s) 
print(t) 
```

Code tersebut digunakan untuk membuat set dan menampilkan isi set.
s = set([3, 4, 9, 10]) artinya membuat set bernama s dari list yang berisi angka 3, 4, 9, dan 10 karena Set tidak memiliki urutan tetap, jadi hasilnya bisa berubah posisi.
t = set("Hello") artinya membuat set dari string "Hello". Set akan mengambil setiap huruf dan menghapus huruf yang sama, seperti Huruf l terdapat dua, tetapi di set hanya akan mencetak satu karena set tidak boleh ada duplikat.

```python
print(s[0])
```

Code tersebut digunakan untuk mencoba mengambil elemen dari set menggunakan indeks, tetapi akan menyebabkan error, karena set tidak memiliki indeks seperti list atau tuple. 
Pada code sebelumnya, s adalah set yang dibuat dari angka 3, 4, 9, dan 10. Seperti yang sebelumnya dikatakan diatas set merupakan tipe data yang tidak berurutan, sehingga setiap elemen tidak memiliki posisi tetap dan tidak bisa diakses menggunakan indeks.
Pada baris print(s[0]), code tersebut mencoba mengambil elemen pada indeks ke-0 dari set s. Namun set tidak mendukung penggunaan indeks, sehingga perintah tersebut tidak bisa dilakukan.

```python
s = set([3, 4, 9, 10])

for x in s:
    print(x)
```

Code tersebut digunakan untuk menampilkan isi set dengan menggunakan perulangan, pada code for x in s:, perintah tersebut digunakan untuk mengambil setiap elemen yang ada di dalam set s satu per satu. Setiap elemen yang diambil akan disimpan sementara ke dalam variabel x.
Kemudian pada baris print(x), nilai yang sedang berada di dalam variabel x akan ditampilkan ke layar. Karena perulangan berjalan sampai semua elemen di dalam set habis, maka semua isi set akan ditampilkan satu per satu.
Hasil yang ditampilkan bisa berbeda urutan setiap kali dijalankan, karena set tidak memiliki urutan tetap.

```python
contoh_set1 = {12, 13, 14, 15} # membuat set bernama contoh_set1 yang berisi angka 12, 13, 14, dan 15
contoh_set2 = {14, 15, 16, 17} # membuat set bernama contoh_set2 yang berisi angka 14, 15, 16, dan 17

print("set 1 =", contoh_set1) # menampilkan tulisan "set 1 =" lalu menampilkan isi dari contoh_set1
print("set 2 =", contoh_set2) # menampilkan tulisan "set 2 =" lalu menampilkan isi dari contoh_set2
```

Karena set bersifat tidak berurutan (unordered) maka output yag dihasilkan akan berbeda-beda atau urutannya acak.

```python
union = contoh_set1 | contoh_set2
print(union)
```

pada baris pertama tanda | digunakan untuk operasi union pada set, union artinya menggabungkan dua set menjadi satu, semua elemen dari contoh_set1 dan contoh_set2 digabung dan jika ada angka yang sama, hanya ditulis satu kali.
output yang dihasilkan adalah {12, 13, 14, 15, 16, 17}
angka 14 dan 15 tidak ditulis dua kali karena set tidak boleh duplikat dan urutan bisa berbeda karena set tidak berurutan.

```python
intersection = contoh_set1 & contoh_set2
print(intersection)
```

Pada baris pertama tanda & digunakan untuk operasi intersection pada set, intersection artinya mencari elemen yang sama dari kedua set dan hanya angka yang ada di contoh_set1 dan contoh_set2 yang akan diambil
output yang dihasilkan {14, 15} karena hanya angka itu yang terdapat di kedua set.

```python
difference = contoh_set1 - contoh_set2
print(difference)
```

pada baris pertama tanda - yang ada diantara contoh_set1 dan contoh_set2 digunakan untuk operasi difference pada set, difference artinya mencari elemen yang hanya ada di contoh_set1 dan tidak ada di contoh_set2.
Maka outputnya 12 dan 13 yang hanya ada di contoh_set1 sedangkan 14 dan 15 tidak ikut karena ada di kedua set.

```python
symmetric_difference = contoh_set1 ^ contoh_set2
print(symmetric_difference)
```

pada baris pertama tanda ^ digunakan untuk operasi symmetric difference pada set. symmetric difference artinya mengambil elemen yang berbeda dari kedua set, elemen yang sama di kedua set tidak ikut diambil
14 dan 15 tidak ikut karena ada di kedua set, yang diambil hanya elemen yang tidak sama.
Makaq outputnya {12, 13, 16, 17}

```python
s = set([3, 4, 9, 10])
t = set("Hello")

print(s)
print(t)
```

Pada baris pertama membuat set s dari list [3, 4, 9, 10]
Pada baris kedua membuat set t dari string "Hello"
setiap huruf akan menjadi elemen set dan jika di cetak huruf yang sama akan dihapus, jadi "l" pada "Hello" tidak akan muncul dua kali.

{9, 10, 3, 4}
{'H', 'e', 'l', 'o'}

```python
t.add('x') 
print(t)
```

Pada add digunakan untuk menambahkan elemen ke dalam set t. Maka huruf 'x' akan ditambahkan ke set, jika elemen belum ada di set, maka akan ditambahkan tetapi jika elemen ada di dalam set tidak akan ditambahkan.
jadi outputnya {'H', 'l', 'e', 'x', 'o'}
kembali lagi urutan bisa berbeda karena set bersifat tidak berurutan.

```python
s.update([10, 37, 42]) 
# update digunakan untuk menambahkan beberapa elemen ke dalam set s
# data yang ditambahkan berasal dari list [10, 37, 42]
# jika ada elemen yang sudah ada di set, maka tidak akan ditambahkan lagi
print(s)
# menampilkan isi set s setelah ditambahkan elemen
# urutan bisa berbeda karena set bersifat tidak berurutan.
```

```python
t.remove('H')
# remove digunakan untuk menghapus elemen dari set t
# elemen 'H' akan dihapus dari set
# jika elemen ada, maka akan berhasil dihapus
# jika elemen tidak ada, akan muncul error
print(t)
# menampilkan isi set t setelah elemen 'H' dihapus
# urutan bisa berbeda karena set bersifat tidak berurutan.
```

```python
a = {
    "username": "labfti",
    "home": "/home/labfti/",
    "uid": 500
}

print(a)
```

a adalah dictionary. Dictionary tersebut memiliki 3 pasangan key dan value:
"username" → "labfti"
"home" → "/home/labfti/"
"uid" → 500
print(a) digunakan untuk menampilkan isi dictionary ke layar.
Program ini membuat dictionary berisi data pengguna lalu mencetaknya ke layar.
output yang dihasilkan:
{'username': 'labfti', 'home': '/home/labfti/', 'uid': 500}

```python
u = a["username"]   
d = a["home"]      

print(u)
print(d)
```

Pada baris pertama artinya kita meminta program mengambil nilai dari key username pada dictionary a, lalu menyimpannya ke variabel u.
Pada baris kedua artinya kita mengambil nilai dari key home pada dictionary a, lalu menyimpannya ke variabel d.
maka outputnya:
labfti
/home/labfti/

```python
a["username"] = "pxl" # Mengubah nilai key "username" dari "labfti" menjadi "pxl"
a["home"] = "/home/pxl" # Mengubah nilai key "home" dari "/home/labfti/" menjadi "/home/pxl"
a["shell"] = "/usr/bin/bash" # Menambahkan key baru yaitu "shell" dengan nilai "/usr/bin/bash"
print(a) 
```

```python
print('username' in a)
#Baris kode ini digunakan untuk mengecek apakah key username ada di dalam dictionary a.
#Kata in berfungsi untuk melakukan pengecekan.
#Jika key username ada, maka outputnya True tetapi jika tidak ada, maka outputnya False.
```

soal bonus 1
apa bedanya list dan tuple?
List digunakan ketika data tersebut masih bisa diubah, misalnya jika ingin menambah, menghapus, atau mengganti data. List ditulis menggunakan tanda kurung siku [ ].
Sedangkan tuple digunakan ketika data tersebut tidak boleh diubah setelah dibuat. Jadi datanya bersifat tetap. Tuple ditulis menggunakan tanda kurung biasa ( ).
Kesimpulannya bahwa list lebih fleksibel karena datanya bisa diubah, sedangkan tuple lebih aman untuk data yang tidak ingin diubah.

Cara Mengubah List ke Dictionary yaitu gunakan fungsi dict().

```python
# Setelah diubah
data = [["Nama","Chatrina"],["NIM","535250096"],["Prodi","Informatika"],["Matakuliah","Numed"]]
hasil = dict(data)
print(hasil)
```
